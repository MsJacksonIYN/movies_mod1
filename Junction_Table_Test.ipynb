```python
create_query = """CREATE TABLE IF NOT EXISTS movie_genres
(   genre_id INTEGER NOT NULL,
    movie_id TEXT    NOT NULL,
    PRIMARY KEY(genre_id, movie_id),
    FOREIGN KEY(genre_id) REFERENCES GENRES(id),
    FOREIGN KEY(movie_id) REFERENCES MOVIES(imdb_id));"""
```

```python
import sqlite3
conn = sqlite3.connect('movies.db')
c = conn.cursor()

def create_table(query):
    c.execute(query)

def close_c_conn():
    c.close()
    conn.close()
```

```python
c.execute('''DROP TABLE movie_genres''')
create_table(create_query)
```

```python
def movie_genre_entry(genre_ids, title_ids):
 
    for i in range(len(genre_ids)):
        c.execute("INSERT INTO movie_genres (genre_id, movie_id) VALUES (?,?)",
          (int(genre_ids[i]), title_ids[i]))
        
    conn.commit()
```
