```python
#URL for getting TMDB id 
   
def get_id_from_api(id_list):
    movie_dict={}
    for movie in id_list:
        i = 0
        url = f'https://api.themoviedb.org/3/find/{movie}?api_key={api_key}&language=en-US&external_source=imdb_id'
            
        response = requests.get(url) 
        movie_dict[movie] = response.json()
        if i%40 == 0:
            time.sleep(2 + random.random())
        i += 1
    
    return movie_dict
        
```

```python
#setting up query to create movies table
create_query = """CREATE TABLE IF NOT EXISTS tmdb_ids
(imdb_id TEXT PRIMARY KEY, tmdb_id TEXT, original_lang TEXT, overview TEXT, popularity REAL, release_date TEXT, title TEXT, vote_avg REAL, vote_count INTEGER)"""
```

```python
#function to create SQLite table. Intakes create_query 
def create_table(query):
    c.execute(query)

#function to close connection
def close_c_conn():
    c.close()
    conn.close()
```

```python
#function to enter data (json file) into the movies SQL table
#data is entered one row at a time
def movie_data_entry(data):
 
    for movie in data:
        imdb_id = movie['index']
        tmdb_id = movie['id']
        lang = movie['original_language']
        rate = movie['overview']
        score = movie['popularity']
        runtime = movie['release_date']
        title = movie['title']
        votes = movie['vote_average']
        year = movie['vote_count']
        
        c.execute("INSERT INTO tmdb_ids (imdb_id, tmdb_id, original_lang, overview, popularity, release_date, title, vote_avg, vote_count) VALUES (?,?,?,?,?,?,?,?,?)",
          (imdb_id, tmdb_id, lang, rate, score, runtime,title,votes,year))
        
       
        
    conn.commit()
```

```python
create_table(create_query)
```

```python
movie_data_entry(data)
```
